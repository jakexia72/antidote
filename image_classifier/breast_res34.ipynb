```python
def predict_score (file_name):
    results = []
    img=open_image(file_name)
    temp = str(learn.predict(img))
    import re
    words = temp.split()
    category = words[1]

    no_score = re.findall(r'\b\d+\b',words[3])[1]
    yes_score = re.findall(r'\b\d+\b',words[4])[1]
    
    results.append(no_score)
    results.append(yes_score)
    return results
```
