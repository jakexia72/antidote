```python
import csv
import pandas as pd
```

```python
df = pd.read_csv("chem_data_colon.csv",delimiter=';')
```

```python
df.head(10)
```

```python
# add this URL to the csv file 
# https://www.ebi.ac.uk/chembl/api/data/image/CHEMBL4116520.svg?engine=indigo
# where anything between image/ and .svg is the ChEMBL ID
urls = []
for s in df["ChEMBL ID"]:
    urls.append("https://www.ebi.ac.uk/chembl/api/data/image/" + str(s)+ ".svg?engine=indigo")
    
df['Images'] = urls
```

```python
df.head(10)
```

```python
df.to_csv("updated.csv")
```

```python
# Next we actually save the SMILES and process them manually through https://pubchem.ncbi.nlm.nih.gov/idexchange/idexchange.cgi
# to get the pubchem id
smiles = df['Smiles'].tolist()
temp1 = smiles[:len(smiles)//2]
temp2 = smiles[len(smiles)//2:]

test1 = temp1[:len(temp1)//2]
test2 = temp1[len(temp1)//2:]
test3 = temp2[:len(temp2)//2]
test4 = temp2[len(temp2)//2:]

with open('SMILES_test1.txt', 'w') as f:
    for item in test1:
        f.write("%s\n" % item)

with open('SMILES_test2.txt', 'w') as f:
    for item in test2:
        f.write("%s\n" % item)

with open('SMILES_test3.txt', 'w') as f:
    for item in test3:
        f.write("%s\n" % item)
          
with open('SMILES_test4.txt', 'w') as f:
    for item in test4:
        f.write("%s\n" % item)
```

```python
# Now we want to write the pubchem CIDs into a list. 

with open("CIDS_test1.txt") as f:
    content = f.readlines()
    # you may also want to remove whitespace characters like `\n` at the end of each line
chemDict = {}

smile_cids = []
for x in content:
    temp = []
    temp = x.split()
    if len(temp) == 1:
        value = None
    else: 
        value = temp[1]

    chemDict[temp[0]] = value
```

```python
with open("CIDS_test2.txt") as f:
    content2 = f.readlines()
    # you may also want to remove whitespace characters like `\n` at the end of each line

for x in content2:
    temp = []
    temp = x.split()
    if len(temp) == 1:
        value = None
    else: 
        value = temp[1]

    chemDict[temp[0]] = value
```

```python
with open("CIDS_test4.txt") as f:
    content4 = f.readlines()
    # you may also want to remove whitespace characters like `\n` at the end of each line

for x in content4:
    temp = []
    temp = x.split()
    if len(temp) == 1:
        value = None
    else: 
        value = temp[1]

    chemDict[temp[0]] = value
```
